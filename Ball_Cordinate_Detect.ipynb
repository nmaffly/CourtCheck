```python
import os
import cv2
from yolov5.detect import detect  # Import detect function from yolov5.detect

# Function to perform object detection on a single image using YOLOv5
def detect_objects_yolov5(image):
    try:
        # Perform object detection using YOLOv5
        results = detect(image)
        # Extract bounding boxes and class labels from detection results
        bboxes = results.pred[0][:, :4].cpu().numpy()
        class_labels = results.pred[0][:, -1].cpu().numpy().astype(int)
        # Draw bounding boxes on the image
        for bbox, class_label in zip(bboxes, class_labels):
            x_min, y_min, x_max, y_max = bbox.astype(int)
            class_name = results.names[int(class_label)]
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
            cv2.putText(image, class_name, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        return image
    except Exception as e:
        print("Error during object detection:", e)
        return None

# Function to load and preprocess a sequence of images from a clip folder
def load_clip_images(clip_folder):
    images = []
    for root, _, files in os.walk(clip_folder):
        for filename in files:
            if filename.endswith(".jpg") or filename.endswith(".png"):
                image_path = os.path.join(root, filename)
                image = cv2.imread(image_path)
                if image is not None:
                    images.append(image)
                else:
                    print("Failed to load image:", image_path)
    return images

# Load images from a single clip folder
clip_folder = "/Users/arnav_a_cherukuthota/Downloads/ASA Tennis Detection/ASA Tennis Object Detection Data/game1"  # Replace with the path to your clip folder
clip_images = load_clip_images(clip_folder)
print("Number of images loaded:", len(clip_images))

# Perform object detection using YOLOv5 on each frame and visualize the results
for image in clip_images:
    detected_image = detect_objects_yolov5(image)
    if detected_image is not None:
        cv2.imshow("Detected Objects", detected_image)
        cv2.waitKey(0)  # Wait for key press to proceed to the next frame
    else:
        print("No detection results for the current frame.")

cv2.destroyAllWindows()
```
