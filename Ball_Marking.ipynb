```python
#Ball Marking Algorithm

import cv2
import pandas as pd
import os

def place_marker(image_path, coordinates, marker_color=(0, 0, 255), marker_size=5):
    """
    Places a marker on the image at the given coordinates.

    :param image_path: Path to the input image.
    :param coordinates: Tuple of (x, y) coordinates where the marker should be placed.
    :param marker_color: Color of the marker in BGR format (default is red).
    :param marker_size: Size of the marker (default is 5).
    :return: Image with the marker.
    """
    # Read the image
    image = cv2.imread(image_path)

    if image is None:
        raise ValueError("Image not found at the provided path.")

    # Unpack the coordinates
    x, y = coordinates

    # Draw the marker on the image
    cv2.circle(image, (x, y), marker_size, marker_color, -1)

    return image

def get_coordinates_from_csv(csv_path, image_name):
    """
    Gets the coordinates for the given image file from the CSV file.

    :param csv_path: Path to the CSV file.
    :param image_name: Name of the image file.
    :return: Tuple of (x, y) coordinates.
    """
    # Read the CSV file
    df = pd.read_csv(csv_path)

    # Filter the DataFrame for the given image file
    row = df[df['file name'] == image_name]

    if row.empty:
        raise ValueError(f"No entry found for the image file {image_name} in the CSV file.")

    # Extract the coordinates
    x = int(row['x-coordinate'].values[0])
    y = int(row['y-coordinate'].values[0])

    return x, y

def process_images_in_folder(folder_path, csv_path, output_folder):
    """
    Processes all images in the specified folder, placing markers based on coordinates from the CSV file.

    :param folder_path: Path to the folder containing images.
    :param csv_path: Path to the CSV file.
    :param output_folder: Path to the folder where output images will be saved.
    """
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # List all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(folder_path, filename)
            try:
                # Get the coordinates from the CSV file
                coordinates = get_coordinates_from_csv(csv_path, filename)

                # Place the marker on the image
                image_with_marker = place_marker(image_path, coordinates)

                # Save the image with the marker
                output_path = os.path.join(output_folder, filename)
                cv2.imwrite(output_path, image_with_marker)

                print(f"Processed and saved: {output_path}")
            except ValueError as e:
                print(e)

def process_clips_in_game_folder(game_folder, output_folder):
    """
    Processes all clips within the specified game folder.

    :param game_folder: Path to the game folder containing clip folders.
    :param output_folder: Path to the folder where output images will be saved.
    """
    # List all subfolders in the game folder
    for clip_folder in os.listdir(game_folder):
        clip_path = os.path.join(game_folder, clip_folder)
        if os.path.isdir(clip_path):
            # Check for the presence of Label.csv in the clip folder
            csv_path = os.path.join(clip_path, 'Label.csv')
            if os.path.exists(csv_path):
                # Define the output path for this clip
                clip_output_folder = os.path.join(output_folder, clip_folder)
                # Process all images in this clip folder
                process_images_in_folder(clip_path, csv_path, clip_output_folder)
            else:
                print(f"No Label.csv found in {clip_path}, skipping this clip.")

def process_all_games(games_folder, output_folder):
    """
    Processes all games, each containing multiple clips.

    :param games_folder: Path to the folder containing game folders.
    :param output_folder: Path to the folder where output images will be saved.
    """
    # List all game folders in the main folder
    for game_folder in os.listdir(games_folder):
        game_path = os.path.join(games_folder, game_folder)
        if os.path.isdir(game_path):
            # Define the output path for this game
            game_output_folder = os.path.join(output_folder, game_folder)
            # Process all clips in this game folder
            process_clips_in_game_folder(game_path, game_output_folder)

def main():
    # Define the path to the main folder containing game folders
    games_folder = '/Users/arnav_a_cherukuthota/Downloads/ASA Tennis Detection/ASA Tennis Object Detection Data'

    # Define the path to the output folder
    output_folder = '/Users/arnav_a_cherukuthota/Downloads/ASA Tennis Detection/ASA Tennis Object Detection Data/BallMarkings'

    # Process all games in the main folder
    process_all_games(games_folder, output_folder)

if __name__ == "__main__":
    main()
```
